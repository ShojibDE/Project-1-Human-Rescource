```python
# Let's replace 'Attritition' , 'overtime' , 'Over18' column with integers before performing any visualizations 
#In these column has simply has yes/no value so we change these to numeric values 

# .apply() take a function and apply it on every single element.
#to know the lambda function see  DS tool box 1 , unit 3 , slide 2, 

employee_df["Attrition"]= employee_df["Attrition"].apply(lambda x : 1 if x=='Yes'  else 0 )
```

```python
#so we check our code work doing write or wrong 
employee_df.head()
```

```python
# Let's replace  'overtime' , 'Over18' column

employee_df["OverTime"]= employee_df["OverTime"].apply(lambda x: 1 if x=="Yes" else 0)
employee_df["Over18"]= employee_df["Over18"].apply(lambda x : 1 if x=="Y" else 0 )
```

```python
# It makes sense to drop 'EmployeeCount' and 'Standardhours' since they do not change from one employee to the other
employee_df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'] , axis= 1, inplace = True )
```

```python
# Let's see how many employees left the company! 
#we divided the Df in to two part. in the :Attrition' column yes 0r 1 means the employee who left the compnay 
left_df= employee_df[employee_df['Attrition']==1]
stayed_df= employee_df[employee_df['Attrition']==0]
```

```python
#total number of employee
print("Total=  ", len(employee_df))
print("Number of employees who left= ", len(left_df))
print("% of employees who left= ",  len(left_df) / len(employee_df), '%' )


print("Number of employees who stayed= ", len(stayed_df))
print("% of employees who stayed= ",  len(stayed_df) / len(employee_df), '%' )
```

```python
X_cat= employee_df[["BusinessTravel" , "Department" , "EducationField" , "Gender" , "JobRole" , "MaritalStatus"]]
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder= OneHotEncoder()
X_cat= onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat= pd.DataFrame(X_cat)
```

```python
print(X_cat.head())
```

```python
X_numerical= employee_df.loc[:, ~employee_df.columns.isin(["BusinessTravel" , "Department" , "EducationField" , "Gender" , "JobRole" , "MaritalStatus"])]
```

```python
X_numerical
```

```python
X_numerical= X_numerical.drop("Attrition", axis = 1) # "Attrition" column is our target column so de drop it from our trin data
```

```python
X_numerical
```

```python
X_all=pd.concat([X_cat , X_numerical], axis= 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=  MinMaxScaler()

X= scaler.fit_transform(X_all)
```

```python
X
```

```python
y= employee_df["Attrition"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size= 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
model= LogisticRegression()
model.fit(X_train , y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("Accuracy {} % ".format(100* accuracy_score(y_pred , y_test)))
```

```python
cm=confusion_matrix(y_pred , y_test)
```

```python
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_pred , y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model.fit(X_train ,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
cm=confusion_matrix(y_pred , y_test)
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_pred , y_test))
```
